A bonus notebook, showcasing inconsistencies in the [Global Transport Costs Dataset for International Trade](https://unctadstat.unctad.org/wds/ReportFolders/reportFolders.aspx?IF_ActivePath=P%2C207045)

It expects [this csv file](https://unctadstat.unctad.org/7zip/US_TransportCosts_Road_Main.csv.7z) to be in the data folder.

```python
import pandas as pd
from pathlib import Path
```

```python
# WARNING: this may take some time, the csv is 1.5GB large
data_folder = Path('../../data')
data = pd.read_csv(data_folder / 'US_TransportCosts_Road_Main_ST202101131545_v1.csv')
# Lots of NA values in that dataset
data.dropna(inplace=True)
data.head()
```

```python
# After some trial & error, the data with France as an origin seems to be more consistent,
# but the overall result still varies wildly in transport costs per 10 000km
france = data[data['Origin Label'] == 'France']

aggregates = france[france['HS2012Product'] == 251710]
aggregates
```

```python
# Transportation costs for France
aggregates['Transport costs per unit per 10 000 km'].mean()
```

```python
# Global average
data[data['HS2012Product'] == 251710]['Transport costs per unit per 10 000 km'].mean()
```
